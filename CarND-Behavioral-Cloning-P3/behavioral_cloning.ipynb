```python
def rgb2YUV(img_set):
    n,row,col,chans = np.shape(img_set)
    
    #normalize RGB 0 to 1
    norm_img_set = np.array(img_set)/255.
    norm_img_set = norm_img_set.astype(np.float32)
    #norm_img_set = cv2.multiply(norm_img_set,np.ones((n,row,col,chans))/255,dtype=cv2.CV_32F) 
    
    mod_img_set = np.array([])
    for img in norm_img_set:
        tImg = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
        mod_img_set = np.append(mod_img_set,tImg)
        
    #reshape
    mod_img_set = np.reshape(mod_img_set,(len(img_set),row,col,chans))
    return mod_img_set

#Assume YUV and just adjust Y channel by random adj
#Must past in labels to output correct label ordering
def luminAugment(img_set,vLabels,yLabels,cLabels,p=0.1,yShape=2): 
    n,row,col,chans = np.shape(img_set)
    adj = np.arange(-.1,.11,.2/10) # want to include .1
    adj = adj[np.abs(adj)>0.0004] # deals with very small numbers
    imgSample = random.sample(range(n),int(n*p))
    
    modImgSet = np.copy(img_set)
    modVLabels = np.copy(vLabels)
    modYLabels = np.copy(yLabels)
    modCLabels = np.copy(cLabels)
    for idx in imgSample:
        tImg = np.copy(img_set[idx])
        tImg[:,:,0] = tImg[:,:,0] + random.sample(list(adj),1)[0]
        modImgSet = np.append(modImgSet,tImg)
        modVLabels = np.append(modVLabels,vLabels[idx])
        modYLabels = np.append(modYLabels,yLabels[idx])
        modCLabels = np.append(modCLabels,cLabels[idx])
    
    #reshape
    modImgSet = np.reshape(modImgSet,(len(modVLabels),row,col,chans))
    modYLabels = np.reshape(modYLabels,(len(modVLabels),yShape))
    
    modImgSet = np.clip(modImgSet,0.,1.) #make sure values are between 0 and 1
    
    return modImgSet.astype(np.float32),modVLabels,modYLabels,modCLabels

def YUV2RGB(img_set):
    n,row,col,chans = np.shape(img_set)
      
    mod_img_set = np.array([])
    for img in img_set:
        tImg = cv2.cvtColor(img, cv2.COLOR_YUV2RGB)
        mod_img_set = np.append(mod_img_set,tImg)
        
    #reshape
    mod_img_set = np.reshape(mod_img_set,(len(img_set),row,col,chans))    
        
    #scale back to 0-255
    mod_img_set = np.uint8(mod_img_set*255.)
    
    return mod_img_set
```

```python
def steeringClassifier(x):
    cla=0
    if abs(x) <= 0.04:
        cla = 0
    elif x>0.04 and x<=0.2:
        cla = 1
    elif x<-0.05 and x >=-0.2:
        cla = -1
    elif x>0.2 and x<=0.45:
        cla = 2
    elif x<-0.2 and x >=-0.45:
        cla = -2
    elif x>0.45 and x<=0.8:
        cla = 3
    elif x<-0.45 and x >=-0.8:
        cla = -3
    elif x>0.8:
        cla = 4
    elif x<-0.8:
        cla = -4
        
    return cla
```

Use trigonometry to calculate adjusted right and left steering angles. If the car is already turing right, then right camer should represent car steering left but not as much compared to center camera going striaght. The left camer would represent car steering harder to the right. 

```python
def calcAdjAngle(c): # c from -1 to 1 represent -25 to 25
    RD = 0.825 # camera mm offset Right
    LD = 0.806 # camera mm offset Left
    cA = math.radians(abs(c)*25.) # center angle, converted to rad
    MD = 18.8956 # = RD/tan(2.5 degree)
    l,r = 0.,0.
    if abs(c)<=0.02: #calc offset when driving straight < 0.5 degree steering
        l = 0.0977
        r = -0.1
    elif c<-0.02:
        x = MD*math.sin(cA)
        if x<LD: 
            lR = math.atan((LD-x)/(MD*math.cos(cA)))
            rR = -math.atan((RD+x)/(MD*math.cos(cA)))
        elif x>LD: # Turn left too hard so x > LD
            lR = -math.atan((x-LD)/(MD*math.cos(cA)))
            rR = -math.atan((x+RD)/(MD*math.cos(cA)))
        else: 
            lR = 0.
            rR = -math.atan((RD+x)/(MD*math.cos(cA)))
        
        l = math.degrees(lR)/25.
        r = math.degrees(rR)/25.
    elif c>0.02:
        x = MD*math.sin(cA)
        if x<RD:
            lR = math.atan((LD+x)/(MD*math.cos(cA)))
            rR = -math.atan((RD-x)/(MD*math.cos(cA)))
        elif x>RD: # Turn right too hard so x > RD
            lR = math.atan((x+LD)/(MD*math.cos(cA)))
            rR = math.atan((x-RD)/(MD*math.cos(cA)))
        else:
            lR = math.atan((LD+x)/(MD*math.cos(cA)))
            rR = 0.
        
        l = math.degrees(lR)/25.
        r = math.degrees(rR)/25.
        
    return l,r
```

```python
# Prepare image for pickling
cImages = np.array([]).astype(np.uint8) # center images
lImages = np.array([]).astype(np.uint8) # left images
rImages = np.array([]).astype(np.uint8) # right images
steering = np.array([]) 
throttle = np.array([])
speed = np.array([])

for csv_row in csv_rows_easy:
    tImg = mpimg.imread(csv_row[0]) # first column is path to center camera image
    tImg = np.uint8(tImg[30:,:,:]) # clip off top 30 pixel
    cImages = np.append(cImages,tImg) 
    
    tImg = mpimg.imread(csv_row[1])
    tImg = np.uint8(tImg[30:,:,:]) 
    lImages = np.append(lImages,tImg)
    
    tImg = mpimg.imread(csv_row[2])
    tImg = np.uint8(tImg[30:,:,:]) 
    rImages = np.append(rImages,tImg)
    
    steering = np.append(steering,float(csv_row[3])) # must convert string to float
    # conbine throttle and break (- throttle)
    throttle = np.append(throttle,float(csv_row[4])-float(csv_row[5])) 
    speed = np.append(speed,float(csv_row[6]))
    
row,col,chans = np.shape(tImg)
    
# reshape 
cImages = np.reshape(cImages,(len(steering),row,col,chans))
lImages = np.reshape(lImages,(len(steering),row,col,chans))
rImages = np.reshape(rImages,(len(steering),row,col,chans))
```

```python
# Sample script used to apply adjusted steering calculation on left and right camera images
# y_data includes steering and throttle so I can keep all recorded data. Throttle will not be used during training.
X_data = np.array([]).astype(np.uint8)
v_data = np.array([])
y_data = np.array([])

# apply adjusted steering calculation for images captured in easy and hard track
for idx in range(len(cImages)):
    lA,rA = calcAdjAngle(steering[idx])
    
    X_data = np.append(X_data,cImages[idx])
    v_data = np.append(v_data,speed[idx])
    y_data = np.append(y_data,[steering[idx],throttle[idx]])
    
    X_data = np.append(X_data,lImages[idx])
    v_data = np.append(v_data,speed[idx])
    y_data = np.append(y_data,[lA,throttle[idx]])
    
    X_data = np.append(X_data,rImages[idx])
    v_data = np.append(v_data,speed[idx])
    y_data = np.append(y_data,[rA,throttle[idx]])
       
row,col,chans = np.shape(cImages[0])

# reshape
X_data = np.uint8(np.reshape(X_data,(len(v_data),row,col,chans)))
y_data = np.reshape(y_data,(len(v_data),2))
```

### Define Generator Specific to Model

```python
def trainDataGenV2(pklPath,p=1.0,steps=80):
    while 1:
        with open(pklPath, mode='rb') as f:
            dataDict = pickle.load(f)
            
        X_data,v_data,y_data,c_data = dataDict['img'],dataDict['aux'],dataDict['y'],dataDict['class']
        step_size = math.floor(len(X_data)/steps)
        
        for idx in range(0,len(X_data),step_size):
            xBatched = rgb2YUV(X_data[idx:int(idx+step_size)])
            vBatched = np.array(v_data[idx:int(idx+step_size)])/30. # scale speed by 1/30
            yBatched = y_data[idx:int(idx+step_size),0] 
            cBatched = c_data[idx:int(idx+step_size)]
            
            X_dataMod,v_dataMod,y_dataMod,c_dataMod = luminAugment(xBatched,vBatched,\
                                                                   yBatched,cBatched,p=p,yShape=1)
            
            y_dataMod2 = np.reshape(np.append(y_dataMod,v_dataMod),(len(v_dataMod),2),order='F')
            
            yield ({'img_in':X_dataMod, 'aux_in':v_dataMod}, {'out':y_dataMod2})
        
def validDataGenV2(pklPath,steps=20):
    while 1:
        with open(pklPath, mode='rb') as f:
            dataDict = pickle.load(f)
            
        X_data,v_data,y_data= dataDict['img'],dataDict['aux'],dataDict['y']
        step_size = math.floor(len(X_data)/steps)
        
        for idx in range(0,len(X_data),step_size):
            xBatched = rgb2YUV(X_data[idx:int(idx+step_size)])
            vBatched = np.array(v_data[idx:int(idx+step_size)])/30. # scale speed by 1/30
            yBatched = y_data[idx:int(idx+step_size),0] 
            
            yBatched2 = np.reshape(np.append(yBatched,vBatched),(len(vBatched),2),order='F')
            
            yield ({'img_in':xBatched, 'aux_in':vBatched}, {'out':yBatched2})
```
